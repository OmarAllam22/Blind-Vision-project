```python
# code for running the recognition project:

import cv2
import numpy as np
import os

# path for our data "different images for each cuurency value"
path = 'E:\\firstpro.py\\images'

mylist = os.listdir(path)
print(mylist)

orb = cv2.ORB_create( nfeatures = 1000 )

deslist = []
for image in mylist:
    kp , des = orb.detectAndCompute(cv2.imread(f'{path}\{image}',0),None)
    deslist.append(des)

def findID(img,deslist,thresh = 15):
    kp2,des2 = orb.detectAndCompute(img,None)
    bf = cv2.BFMatcher()
    matchlist=[]
    final_value= -1
    try:
        for des in deslist:
            matches = bf.knnMatch(des,des2,k=2)
            good=[]
            for m, n in matches:
                if m.distance <0.75*n.distance:
                    good.append([m])
            matchlist.append(len(good))
    except: pass
    
    if len(matchlist) != 0:
        if max(matchlist) > thresh:
            final_value = matchlist.index(max(matchlist))
    return final_value
    #print(matchlist)


cap = cv2.VideoCapture(0)

while True:
    success, img2 = cap.read()
    imgOriginal = img2.copy()
    img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)
    id = findID(img2,deslist)
    if id != -1:
        cv2.putText(imgOriginal,mylist[id],(50,50),cv2.FONT_HERSHEY_COMPLEX,1, (0,0,255),2)
        tts=gTTS(text='ا',lang='ar')
        if mylist[id]== 'fivepounds.jpg' or mylist[id]== 'fivepounds1.jpg' or mylist[id]=='fivepounds2.jpg'  or mylist[id]== 'fivepounds3.jpg' or mylist[id]== 'fivepounds5.jpg':
            playsound('audio5.mp3')
        elif mylist[id] == 'fiftypounds.jpg' or mylist[id]=='fiftypounds1.jpg' or mylist[id]=='fiftypounds2.jpg' or mylist[id]=='fiftypounds3.jpg' or mylist[id]=='fiftypounds4.jpg'  or mylist[id]=='fiftypounds7.jpg' or mylist[id]=='fiftypounds8.jpg' :
             playsound('audio50.mp3')
        elif mylist[id] == 'hunderedpounds.jpg' or mylist[id]== 'hunderedpounds1.jpg'  or mylist[id]== 'hunderedpounds2.jpg'  or mylist[id]== 'hunderedpounds3.jpg'   or mylist[id]== 'hunderedpounds5.jpg'  or mylist[id]== 'hunderedpounds6.jpg'  or mylist[id]== 'hunderedpounds7.jpg'  or mylist[id]== 'hunderedpounds8.jpg'   or mylist[id]== 'hunderedpounds10.jpg' :
            playsound('audio100.mp3')
        elif mylist[id]=='tenpounds.jpg' or mylist[id]=='tenpounds1.jpg' or mylist[id]=='tenpounds2.jpg' or mylist[id]== 'tenpounds3.jpg' or mylist[id]=='tenpounds4.jpg' or mylist[id]=='tenpounds5.jpg' or mylist[id]=='tenpounds6.jpg' or mylist[id]=='tenpounds7.jpg'   or mylist[id]=='tenpounds9.jpg'  or mylist[id]=='tenpounds11.jpg' or mylist[id]=='tenpounds12.jpg'  or mylist[id]=='tenpounds14.jpg' or mylist[id]=='tenpounds15.jpg' or mylist[id]=='tenpounds16.jpg' or mylist[id]=='tenpounds17.jpg' or mylist[id]=='tenpounds18.jpg' or mylist[id]=='tenpounds19.jpg' or mylist[id]=='tenpounds20.jpg'  :
             playsound('audio10.mp3')
        elif   mylist[id]== 'tewentypounds.jpg' or mylist[id]=='tewentypounds1.jpg'   or mylist[id]== 'tewentypounds2.jpg'   or mylist[id]== 'tewentypounds5.jpg'  or mylist[id]== 'tewentypounds6.jpg'  or mylist[id]== 'tewentypounds7.jpg'  or mylist[id]== 'tewentypounds8.jpg'  or mylist[id]== 'tewentypounds9.jpg'  :
            playsound('audio20.mp3')
        elif mylist[id] =='twohunderedpounds.jpg' or mylist[id] =='twohunderedpounds1.jpg' or mylist[id]=='twohunderedpounds2.jpg' or mylist[id]=='twohunderedpounds3.jpg' or mylist[id]=='twohunderedpounds5.jpg' or mylist[id]=='twohunderedpounds6.jpg' or mylist[id]=='twohunderedpounds7.jpg' or mylist[id]=='twohunderedpounds8.jpg' :
            playsound('audio200.mp3')
        elif mylist[id] == 'halfpound.jpg' or mylist[id] == 'halfpound1.jpg' or mylist[id]=='halfpound2.jpg' or mylist[id]=='halfpound5.jpg' or mylist[id]=='halfpound6.jpg' or mylist[id]=='halfpound7.jpg' or mylist[id]=='halfpound8.jpg':
            playsound('audio0.5.mp3')    
        elif  mylist[id] == 'onepound.jpg' or mylist[id] == 'onepound1.jpg' or mylist[id]=='onepound2.jpg' or mylist[id]=='onepound4.jpg' or mylist[id]=='onepound5.jpg' or mylist[id]=='onepound6.jpg' or mylist[id]=='onepound7.jpg' or mylist[id]=='onepound8.jpg'  :
            playsound('audio1.mp3') 
       
          
    cv2.imshow("image",imgOriginal)
    
    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
